### 4.1 Halveringsmetoden -algoritmisk tenkning
Hvordan kan vi finne nullpunkter dersom vi ikke kan lse en likning ved hjelp av det vi har lrt?

I disse ktene med python-programmering har du lrt litt om algoritmisk tenkning og litt om lkker. Vi skal se hvordan vi kan kombinere dette slik at vi kan lse likninger i python. 

Tenk deg at vi tar utgangspunkt i et intervall $[a,b]$, se figuren under, som er slik at funksjonsverdien har forskjellig fortegn i endepunktene til intervallet (p ulik side av x-aksen). Da vet du at nullpunktet til funksjonen m vre i dette intervallet. Midtpunktet,$m_{1}$, beregnes og vi lager et nytt intervall ved  flytte enten a til $m_{1}$ eller b til $m_{1}$ slik at det nye intervallet $ [a_{2},b_{2}]$ ogs inneholder nullpunktet. Vi har n ftt et smalere intervall. Vi halverer intervallet igjen og ser p halvdelen som har forskjellig fortegn p funksjonsverdiene i endepunktet til det nye intervallet $ [a_{3},b_{3}]$. Dette fortsetter vi med til endepunktene er s nr hverandre som vi nsker. Endepunktene er da en tilnrming til nullpunktet.

Denne metoden finner bare et nullpunkt. Vi m derfor dele opp i to ulike intervall og kjre metoden to ganger om vi skal finne to nullpunkt for samme funksjon.

Vi m ogs "gjette" riktig intervall som nullpunktet befinner seg innenfor

### 4.1.1 Eksempel: finne nullpunkt til en liner funksjon.

```python
# Laget av Knut Skrindo, Programmering i matematikk (noe modifisert)
#Vi skal lse likningen 3x-4=0. Vi legger merke til at med f(x)=3x-4 har f(-2) og f(2) ulikt fortegn.
#Velger nedre grenser ut fra dette.
def f(x):
    return 3*x - 4 

nedregrense = -2
vregrense = 2
presisjon = .001

a = nedregrense
b = vregrense
m = (a + b)/2
feil = False

while b - a > presisjon:
   if f(a) * f(m) < 0:
      b = m
   elif f(b) * f(m) < 0:
      a = m
   elif f(m) == 0:
      break
   else:
      feil = True
      break
   m = (a + b)/2

if feil:
    print(f"f({a}) og f({b}) er p samme side av x-aksen.")
    print(f"Jeg finner ingen lsning.")
else:
    print(f"Funksjonen har et nullpunkt i x = {m:.3f}.")
```

### Eksempel 6.1.5 Finne nullpunkt til logaritmefunksjoner med ett nullpunkt.
Vi ser p likningen: $lg(x)+3x-4=0$

Kan vi lse denne likningen?
Svaret er egentlig at vi ikke kan lse denne likningen ved hjelp av det vi har lrt i 1T uten programmering. Det er derimot mulig  bruke halveringsmetoden.

```python
import numpy as np
def f(x):
    return np.log10(x)+3*x - 4 

nedregrense = np.exp(-15)
vregrense = 2
presisjon = .0001

a = nedregrense
b = vregrense
m = (a + b)/2
feil = False

while b - a > presisjon:
   if f(a) * f(m) < 0:
      b = m
   elif f(b) * f(m) < 0:
      a = m
   elif f(m) == 0:
      break
   else:
      feil = True
      break
   m = (a + b)/2

if feil:
    print(f"f({a}) og f({b}) er p samme side av x-aksen.")
    print(f"Jeg finner ingen lsning.")
else:
    print(f"Funksjonen har et nullpunkt i x = {m:.4f}.")
```

### 6.2 Finne flere nullpunkter ved halveringsmetoden
Vi m gjenta halveringsmetoden flere ganger med forskjellige intervaller dersom vi skal finne flere nullpunkt. Vi kan da "gjette" startverdi og sluttverdi til vi finner nullpunktene. Ofte kan det vre vanskelig  avgjre hvor mange nullpunkt en funksjon vil ha. Ved  tegne grafen til funksjonen vil vi kunne se hva som er fornuftige innverdier i halveringsmetoden. Kjr programmet under!

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2,3,100) #Her velger vi x-verdier fra -2 til 3, jevnt forledt p 100 punkter.
z=0*x
def f(x): #definerer funksjonen f
  return( np.log10(x**4)+3*x +5)
plt.plot(x,f(x),label=r'$lg(x^4)+3\cdot x+5$') #tegner grafen, markrer vises p hvert punkt (kan oftest droppes).
plt.plot(x,z)
plt.xlabel('x') #aksetittel x-akse
plt.ylabel('y') #aksetittel y-akse
plt.legend() #navn p funksjon
```

Vi ser av plottet at funksjonen har to nullpunkt. Vi velger start -og sluttverdier utfra grafen.

```python
import numpy as np
def f(x):
    return ( np.log10(x**4)+3*x +5)

nedregrense = -1
vregrense = 0.01
presisjon = .0001

a = nedregrense
b = vregrense
m = (a + b)/2
feil = False

while b - a > presisjon:
   if f(a) * f(m) < 0:
      b = m
   elif f(b) * f(m) < 0:
      a = m
   elif f(m) == 0:
      break
   else:
      feil = True
      break
   m = (a + b)/2

if feil:
    print(f"f({a}) og f({b}) er p samme side av x-aksen.")
    print(f"Jeg finner ingen lsning.")
else:
    print(f"Funksjonen har et nullpunkt i x = {m:.4f}.")
```
